```python
#Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
from pathlib import Path
crime_data_to_load=Path("/Users/Nico/Documents/Resources/Starter_Code/crime-data-from-2010-to-present.csv")
crime_data_df = pd.read_csv(crime_data_to_load, parse_dates=["Date Reported","Date Occurred"])
```

```python
res=[]
for row in crime_data_df.iterrows():
    date = (row[1]["Date Occurred"])
    year,month,day,time=(date.year,date.month,date.day,row[1]["Time Occurred"])
    time=str(time)
    hour,minute=time[:2],time[2:]
    d = datetime.datetime(year=year, month=month, day=day, hour = int(hour) ,minute = int(minute))
    res.append(d)
    
```

```python
crime_data_df["hour occurred"] = crime_data_df["Time Occurred"].apply(lambda x: int(str(x)[:2]))
```

```python
crime_data_df.loc[:, "datetime occurred"]=res 
```

```python
#Display bar graph of the Median of Victim Age
median=crime_data_df['Victim Age'].groupby(crime_data_df['Area Name']).median()
median_stat=pd.DataFrame({"Median of Victim Age":median})

plot_pandas=median_stat.plot.bar(color='b')
plt.xlabel("Area Name")
plt.ylabel("Victim Age")
plt.title("Median of Victim's Age per Area Name")
```
